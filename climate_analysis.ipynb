```python
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
import pandas as pd
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
from sqlalchemy.orm import Session
# Declare a session
session = Session(bind=engine)
```

```python
Base = automap_base() # Declare a Base using `automap_base()`
Base.prepare(engine, reflect=True) # Use the Base class to reflect the database tables
Base.classes.keys() # Print all of the classes mapped to the Base
```

```python
Measurement = Base.classes.measurement
session = Session(engine)
first_row = session.query(Measurement).first()
print(first_row.__dict__)
print('-----------------------------')
for row in session.query(Measurement.prcp, Measurement.tobs).limit(15).all():
    print(row)
```

```python
Station = Base.classes.station

r_s = session.query(Station.id, Station.elevation, Station.latitude,Station.longitude,Station.name,Station.station).\
    order_by(Station.station.desc()).all()

r_df = pd.DataFrame(r_s, columns=['id', 'elevation', 'latitude','longitude','name','station'])
m_df = pd.DataFrame(results, columns=['id', 'date', 'prcp','station','tobs'])


######### Joins
hawaii_df = pd.merge(r_df, m_df, on="station", how="left")
hawaii_df.set_index('station', inplace=True, )
del hawaii_df['id_x']
del hawaii_df['id_y']

################################################# Count obvs. per stations for all 

station_obs_count=hawaii_df.groupby('station').count()
station_obs_count=station_obs_count['tobs']
stations_obs_count_all = pd.DataFrame({
                                    "Number of Observations":station_obs_count,
                                     })
stations_obs_count_all = stations_obs_count_all.sort_values('Number of Observations',ascending=False)

################################################# Count obvs. per stations for a 12 months period 
# Number of obs for a one year period (2016-17)
mask = (hawaii_df['date'] > '2016-08-23') & (hawaii_df['date'] <= '2017-08-23')
hawaii_df_oneyear = hawaii_df.loc[mask]

station_obs_count_oneyr=hawaii_df_oneyear.groupby('station').count()
station_obs_count_oneyr=station_obs_count_oneyr['tobs']
stations_obs_count_oneyr_final = pd.DataFrame({
                                    "Number of Observations":station_obs_count_oneyr,
                                     })
stations_obs_count_oneyr_final = stations_obs_count_oneyr_final.sort_values('Number of Observations',ascending=False)

################################################# Histogram
hawaii_df_oneyear['date'] = pd.to_datetime(hawaii_df_oneyear['date'])
hawaii_df_oneyear['Year, Month'] =hawaii_df_oneyear.date.apply(lambda x: str(x)[:7])

# month_bins = ['2016-08','2016-09','2016-10','2016-11','2016-12','2017-01','2017-02','2017-03','2017-04','2017-05',
#              '2017-06','2017-07','2017-08',]
# month_bins_labs = ['Aug.16','Sep.16','Oct.16','Nov.16','Dec.16','Jan.17','Feb.17','Mar.17','Apr.17','May.17','Jun.17',
#                   'Jul.17','Aug.17']
# month_hist=pd.cut(stations_obs_count_oneyr_final["Number of Observations"].unique(), month_bins, labels=month_bins_labs)

station_obs_count_months=hawaii_df_oneyear.groupby('Year, Month').count()
station_obs_count_months = station_obs_count_months['tobs']
station_obs_count_months_final = pd.DataFrame({
                                    "Number of Observations":station_obs_count_months
                                     })


#####################################################################################################################
print('**********************')
print('Queries')
print('**********************')
print('Number of Observations per Stations')
print('                      ')
print(stations_obs_count_all)
print('                      ')
print('Station USC00519281 has the highest number of observation at 2772 counts')
print('---------------------')
print('Number of Observations per Stations')
print('                      ')
print(stations_obs_count_oneyr_final)
print('                      ')
print('Station USC00519397 has the highest number of observation at 358 counts')

print(station_obs_count_months_final)

data = [
    go.Bar(
        x=station_obs_count_months_final.index,
        y=station_obs_count_months_final['Number of Observations']
    )
]

py.iplot(data)
```
